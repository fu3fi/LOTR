```python
import pandas as pd 
import networkx as nx 
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from pprint import pprint
import scipy.stats
import numpy as np 
import pylab 
import scipy.stats as stats
from scipy.stats import norm


edges = pd.read_csv(r'datasets-edu/marvel/marvel-unimodal-edges.csv')
nodes = pd.read_csv(r'datasets-edu/marvel/marvel-unimodal-nodes.csv')

# строим граф
G = nx.from_pandas_edgelist(edges, 'Source', 'Target')

def histogram(data, cumulative=False, x_label = "", y_label = "", title = ""):
    _, ax = plt.subplots()
    ax.hist(data, cumulative = cumulative, color = '#539caf')
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)

degrees = (list(dict(G.degree()).values()))
pprint(np.histogram(degrees))
```

```python
# в обычных координатах 
stats.probplot(degrees, dist='norm', plot=pylab)
pylab.show()
```

```python
# в логарифмических координатах
stats.probplot(degrees, dist='loggamma', sparams=(1,1), plot=pylab)
pylab.show()
```

```python
# еще один пример построения в обычных координатах
s = pd.Series(degrees)
ax = s.plot.kde()
```

```python
# вызывается функции для построения в обычных координатах
histogram(degrees)
```
