В силу неустоявшегося характера научного направления в нём присутствует сборная терминология:
* network = graph = граф = сеть;

* links = edges, relations = связи в графе (relations – в экономике);
* clusters = communities = сообщества = группы узлов, которые теснее связаны между собой, чем с остальным графом.

````python
import networkx as nx
import pandas as pd
import numpy as np
from networkx.algorithms import bipartite


# Множество работников
employees = set(['Pablo',
                 'Lee',
                 'Georgia',
                 'Vincent',
                 'Andy',
                 'Frida',
                 'Joan',
                 'Claude'])

# Множество фильмов
movies = set(['The Shawshank Redemption',
              'Forrest Gump',
              'The Matrix',
              'Anaconda',
              'The Social Network',
              'The Godfather',
              'Monty Python and the Holy Grail',
              'Snakes on a Plane',
              'Kung Fu Panda',
              'The Dark Knight',
              'Mean Girls'])


# Вы можете иhttp://localhost:8888/files/Desktop/PZ-1%5BStudent%5D/pic/undirected.png?1599834750753спользовать следующую функцию для отображения графа
def plot_graph(G, weight_name=None):
    '''
    G: a networkx G
    weight_name: name of the attribute for plotting edge weights (if G is weighted)
    '''
    %matplotlib notebook
    import matplotlib.pyplot as plt
    
    plt.figure()
    pos = nx.spring_layout(G)
    edges = G.edges()
    weights = None
    
    if weight_name:
        weights = [int(G[u][v][weight_name]) for u,v in edges]
        labels = nx.get_edge_attributes(G,weight_name)
        nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
        nx.draw_networkx(G, pos, width=weights);
    else:
        nx.draw_networkx(G, pos);
````

Используя NetworkX, загрузите двудольный граф из `Movie_Choices.txt` и верните этот граф.

## Задание 2

Используя граф из предыдущего задания, добавьте атрибуты узлов `'type'`. Фильмы имеют значение `'movie'`, а сотрудники имеют значение`'employee'`. И верните этот граф.

*Эта функция должна возвращать граф networkx с атрибутами узла `{'type': 'movie'}` или `{'type': 'employee'}`*

````python
def answer_two(buff):
    G=nx.Graph()
    new_buff = buff.split('\n')[1:-1]
    for employee_and_movie in new_buff:
        name_employee, name_movie = employee_and_movie.split('\t')
        G.add_node(name_employee, type="employee")
        G.add_node(name_movie, type="movie")
        G.add_edge(name_employee, name_movie)
    return G
````

````python
G = answer_two(buff)
G.nodes(data=True)
````

## Задание 3

Найдите взвешенную проекцию графа из answer_two (двудольный граф), которая говорит нам, сколько фильмов объединяют разные пары сотрудников.

*Эта функция должна возвращать взвешенный граф.*

Пример создания двудольного графа

```python
from networkx.algorithms import bipartite
B = nx.Graph() 
B.add_nodes_from(['A','B','C','D', 'E'], bipartite=0) 
B.add_nodes_from([1,2,3,4], bipartite=1)
B.add_edges_from([('A',1), ('B',1), ('C',1), ('C',3), ('D',2), ('E',3), ('E', 4)])
```

<img src="./pic/bipartite.png" width=200>

````python
import itertools
from collections import Counter

def get_emploees(buff):
    new_buff = buff.split('\n')[1:-1]
    return set(employee_and_movie.split('\t')[0] for employee_and_movie in new_buff)

def get_movies(buff):
    new_buff = buff.split('\n')[1:-1]
    return set(employee_and_movie.split('\t')[1] for employee_and_movie in new_buff)

def answer_three(buff):
    movies = get_movies(buff)
    pairs_list = list()
    G = answer_two(buff)
    for movie in movies:
        pairs_list.extend(list(itertools.combinations([neighbor for neighbor in G.neighbors(movie)], 2)))
    T = nx.Graph() 
    for nodes, weight in dict(Counter(pairs_list)).items():
        T.add_edge(*nodes, weight=weight)
    return T
````

## Задание 4

Предположим, вы хотите узнать, нравятся ли людям с высоким показателем взаимоотношений одинаковые фильмы.

Найдите корреляцию Пирсона (используя `DataFrame.corr ()`) между оценками отношений сотрудников (`Relationships.txt`) и количеством общих фильмов(`Movie_Choices.txt`). Если у двух сотрудников нет общих фильмов, их следует рассматривать как 0, а не как отсутствующее значение, и их следует включать в расчет корреляции.

*Эта функция должна возвращать число с плавающей точкой.*

Коэффициент корреляции Пирсона (r-Пирсона) применяется для исследования взаимосвязи двух переменных, измеренных в метрических шкалах на одной и той же выборке. Он позволяет определить, насколько пропорциональная изменчивость двух переменных. 

Данный коэффициент разработали Карл Пирсон, Фрэнсис Эджуорт и Рафаэль Уэлдон в 90-х годах XIX века. Коэффициент корреляции изменяется в пределах от минус единицы до плюс единицы. 

Коэффициент корреляции r-Пирсона характеризует существование линейной связи между двумя величинами. Если связь криволинейная то он не будет работать. 

Чтобы приступать к расчетам коэффициента корреляции r-Пирсона необходимо выполнение следующих условий: 
1. Исследуемые переменные X и Y должны быть распределены нормально. 
2. Исследуемые переменные X и Y должны быть измерены в интервальной шкале или шкале отношений. 
3. Количество значений в исследуемых переменных X и Y должно быть одинаковым. 

При расчете коэффициент линейной корреляции Пирсона используется специальная формула. Величина коэффициента корреляции варьируется от 0 до 1.

````python
with open("Relationships.txt", "r") as open_file:
    relationships_buff = open_file.read()

def get_weight_edge(G, node_1, node_2):
    return 0 if G.get_edge_data(node_1, node_2) is None else G.get_edge_data(node_1, node_2)['weight']

def answer_four(buff, relationships_buff):
    G = answer_three(buff)
    G_list = [get_weight_edge(G, *pair) for pair in itertools.combinations(sorted(list(set(get_emploees(buff))), key=lambda x: x[0]), 2)]
    T_list = [int(con_info.split('\t')[-1]) for con_info in relationships_buff.split('\n')[:-1]]
    return pd.DataFrame(zip(G_list, T_list)).corr()[0][1]
````

````python
answer_four(buff, relationships_buff)
````
