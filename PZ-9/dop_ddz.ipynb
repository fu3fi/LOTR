# Метод Клаузета, Ньюмана, Мура

<b>Fastgreedy</b> — метод, основанный на жадной оптимизации функции модулярности.

## Модулярность

Самой популярной и общепризнанной мерой качества для сравнения разбиений является значение <b>модулярности (modularity)</b>. Функционал был предложен Ньюманом и Гирваном в ходе разработки алгоритма кластеризации вершин графа. Модулярность — это скалярная величина из отрезка `[−1, 1]`, которая количественно описывает неформальное определение структуры сообществ:

$Q=\frac{1}{2m} \sum_{i, j}(A_{ij} - \frac{d_i d_j}{2m})\delta(C_i, C_j),$

где $A$ — матрица смежности графа, $A_{ij}$ — $(i, j)$ элемент матрицы, $d_i$ — степень $i$ вершины графа, $C_i$ — метка вершины (номер  сообщества, к которому относится вершина), $m$ — общее количесство ребер в графе. $\delta(C_i C_j)$ — дельта функция: равна единице, если $C_i = C_j$, иначе нулю.

Задача поиска выделения сообществ в графе сводится к поиску таких $C_i$, которые максимизируют значение модулярности.

Формула имеет множество обобщений, например, для взвешенных графов, под $A_{ij}$ понимается вес ребра, соединяющего вершины $i$ и $j$, а $m=\frac{1}{2}\sum a_{ij}$.

### Достоинства модулярности

Модулярность достаточно просто интерпретируется. Ее значение равно разности между долей ребер внутри сообщества и ожидаемой доли связей, если бы ребра были размещены случайно.

Модулярность возможно эффективно пересчитывать при небольших изменениях в кластерах. Например, если добавить изолированную вершину в кластер $C$, то изменение функционала для взвешенного графа можно посчитать как

$\Delta Q = [\frac{\sum_{in} + d_{i,in}}{2m} - (\frac{\sum_{tot} + d_i}{2m})^2] - [\frac{\sum_{im}}{2m} + (\frac{\sum_{tot}}{2m})^2 - (\frac{d_i}{2m})^2]$,

где $\sum_{in}$ — сумма весов ребер внутри сообщества $C$, $\sum_{tot}$ — сумма весов всех ребер инцидентных вершинам из $C$, $d_i$ —  сумма весов ребер инцидентных вершине $i$, $d_{i,in}$ — сумма весов ребер соединяющих вершину $di$ и вершину из $C$, $m$ — сумма весов всех ребер.

### Недостатки модулярности

Функционал <b>не является непрерывным</b>, и задача его оптимизации — <b>дискретная</b>. Для поиска глобального оптимума используют приближенные схемы. Некоторые из них действительно оптимизируют значение функционала, другие же по значению модулярности выбирают наилучшее решение из найденных, то есть без гарантий локальной оптимальности решения.

У данного функционала существует <b>проблема разрешающей способности</b> (грубо говоря, функционал не видит маленькие сообщества). Эта проблема решается путем использования модифицированного функционала, который сохраняет все достоинства и добавляет параметр масштаба.

## Fastgreedy

Алгоритм заключается в жадной оптимизации модулярности. Алгоритм работает по следующей схеме:
1. Инициализация сообществ в каждой вершине.
2. Объединение сообществ, в результате которого максимальным образом увеличивается модулярность.
3. Выход, если дальнейшее увеличение значения функционала невозможно.

Обычно на втором шаге перебираются не все пары сообществ, а только те, между вершинами которых существуют связи.

### Достоинства алгоритма

Данный алгоритм <b>является быстрым и вычислительно простым</b>, что <b>позволяет применять его для больших графов</b>, таких как граф интернет-ресурсов. Так же в него <b>легко добавить априорную информацию о составе кластеров</b>, например, если мы знаем, что какие-то конкретные вершины должны лежать в одном кластере.

### Недостатки алгоритма

Метод содержит в себе все недостатки, свойственные жадным методам, и сходится не к самому лучшему решению. <b>Часто метод порождает одно большое сообщество с большинством вершин графа в нем и множество маленьких.</b>

### Сложность алгоритма

$O(mn), \space m=|E|,\space n=|V|,\space E - множество\space ребер\space графа,\space V - множество\space вершин\space графа.$

## Функция, реализующая алгоритм

<h3>NetworkX</h3>

```python
IGraph.community_fastgreedy(weights=None)
```
Находит структуру сообщества графа в соответствии с алгоритмом Клаузета, Ньюмана, Мура, основанным на жадной оптимизации модулярности.

<b>Параметры</b>:
- weights - имя атрибута ребра или список, содержащий веса ребра.

<b>Возвращаемые значения</b>:
кортеж со следующими элементами
1. Список слияний.
2. Оценка модулярности перед каждым слиянием.


<h3>iGraph</h3>

```python
NetworkX.greedy_modularity_communities(G, weight=None)
```
Находит сообщества в графе, используя жадную максимизацию модульности Клаузета-Ньюмана-Мура. В настоящее время этот метод поддерживает класс Graph и не учитывает веса ребер.

<b>Параметры</b>:
- G - NetworkX граф
- weights - имя атрибута ребра или список, содержащий веса ребра.

<b>Возвращаемые значения</b>: наборы узлов, по одному для каждого сообщества.


## Тестовый граф

Узлами данного графа являются друзья из Вконтакте исходного пользователя, для которого мы выгружали данные в блоке выше. 

Ребро между двумя узлами добаляется тогда и только тогда, когда между данными пользователями есть отношение дружбы Вконтакте, при этом, если `user2` есть в друзьях у `user1`, а `user1` - в друзьях у `user2`, то вес ребра равен `2`, в ином случае вес равен `1`.

````python
import networkx as nx
import json
import matplotlib.pyplot as plt # For graphs
from networkx.algorithms.community import greedy_modularity_communities
import igraph as ig
````

### NetworkX

````python
with open('graph1.json', "r") as f:
    graph_data1 = json.loads(f.read())

G = nx.Graph()
G.add_nodes_from(graph_data1['nodes'])
G.add_weighted_edges_from(graph_data1['edges'])

plt.figure(figsize=(9,6))
nx.draw_kamada_kawai(G, node_size=50, node_color="#FF6347", edge_color="#808080")
````

````python
def nx_draw(G, size=(12, 8)):
    '''Draw networkx graph with communities'''
    c = greedy_modularity_communities(G)
    rgb = [[0, 0, 0.5], [0.5,0 , 0.5], [0.5, 0, 0], [1, 0, 0], [0, 0, 1], [0, 1, 0], [1, 0, 0.5], [0.5, 0, 1], [0, 1, 1], [1, 1, 0], [1, 0, 1], [0.5, 0.5, 0], [0, 0.5, 0.5], [0, 0.5, 0]]
    data = {}
    if len(c) > len(rgb):
        return "Color error"
    for group_index in range(0, len(c)):
        color = rgb[group_index]
        data['group_'+str(group_index)] = {'color' : color, 'count' : len(c[group_index]), 'elements' : list(c[group_index]), 'coverage' : nx.algorithms.community.quality.coverage(G, c), 'perfomance' : nx.algorithms.community.quality.performance(G, c)}
        for elem in c[group_index]:
            G.nodes[elem]['color'] = tuple(color)
            
    plt.figure(figsize=size)
    colors = [G.nodes[elem]['color'] for elem in G.nodes()]
    nx.draw_kamada_kawai(G, node_size=50, node_color=colors, edge_color="#909090")
    return data
````

````python
nx_draw(G)
````

### iGraph

````python
g = ig.Graph.from_networkx(G)
````

````python
def find_cluster(vertex, clusters):
    for i in range(len(clusters)):
        for vs in clusters[i].vs:
#             print(vs)
            if vs['_nx_name'] == vertex['_nx_name']:
                return i
    return -1

def idraw(igraph):
    '''Draw igraph with communities'''
    color_list = ['#FF0000','#0000FF','#000080','#FF00FF','#FFFF00','#00FF00','#00FFFF','#008000','#FFC0CB','#FF1493','#FF4500', '#BDB76B', '#00FA9A', '#696969']
    community = ig.Graph.community_fastgreedy(igraph).as_clustering().subgraphs()
    print(
        {
            'Edges_count' : igraph.ecount(),
            'Nodes_count' : igraph.vcount(),
            'Graph_radius' : igraph.radius(),
            'avg_node_degree' : sum([a for a in igraph.indegree()]) / igraph.vcount()
        }
    )
    return ig.plot(igraph, vertex_size=7, layout=igraph.layout("kk"), vertex_color=[color_list[x] for x in [find_cluster(vs, community) for vs in igraph.vs]])
````

````python
idraw(g)
````
