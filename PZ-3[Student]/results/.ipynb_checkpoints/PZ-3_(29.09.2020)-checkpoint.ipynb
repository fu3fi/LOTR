### 1. Исследование явления фазового перехода для модели ER 
1. Написать функцию, вычисляющую оценки значений характеристик графа в модели ER_n_p:
   ожидаемое (в среднем) количество рёбер, ожидаемая (в среднем) степень узла, оценка диаметра графа.
2. Написать функцию, вычисляющую характеристики реального графа (количество вершин, рёбер, средняя степень узла, коэффициент кластеризации, диаметр, средняя длина пути, отклонения в % реальных значений от теоретических значений).
3. Написать функцию визуализации графа (узлы с метками). Подобрать раскладку графа, наиболее наглядно представляющую устройство графа (возможно сделать раскладку аргументом функции).
4. Написать функцию, вычисляющую максимальный связный подграф данного графа (nx.strongly_connected_components). Если количество вершин этого подграфа  не меньше 3/4 количества вершин данного графа, подграф выводится на экран и представляются его характеристики.
5. Написать функцию, строящую гистограмму частот степеней вершин.
6. Используя функцию nx.gnp_random_graph,  сгенерировать случайный граф с параметрами     
   n = 100, p = 0,001. Применить к нему указанные выше функции. Увеличивая значение параметра p, найти его минимальное значение (с точностью до 1/(10n)), при котором проявляется эффект фазового перехода (связный подграф с 3/4 вершин). Распечатать этот подграф и его характеристики.
7. Повторить п.6 для параметров n = 1000, p = 0,0001; n = 10000, p = 0,00001.    
8. Повторить пп.6,7 для функций nx.erdos_renyi_graph, nx.binomial_graph.
9. Используя функцию nx.gnm_random_graph,  сгенерировать случайный граф с параметрами     
   n = 1000, m = 1000. Применить к нему указанные выше функции. Увеличивая значение параметра m, найти его минимальное значение, при котором проявляется эффект фазового перехода (связный подграф с 3/4 вершин). Распечатать этот подграф и его характеристики.
10. Сравнить результаты генерации случайных графов по разным функциям модели ER. Наблюдаете ли Вы какие-либо закономерности? Можно ли сделать выводы о предпочтительности тех или иных функций для моделирования случайных графов?
