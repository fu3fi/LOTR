```python
import numpy as np
import networkx as nx
from collections import Counter
from matplotlib import pylab as plt
from math import copysign
from math import ceil
import time

γ = 0.5772156649015328606065120
        
def plot_chart(title="", x="", y="", log=False, *charts):
    plt.figure(figsize=(8,6))
    plt.tight_layout()
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid()
    if log == True:
        plt.xscale("log")
        plt.yscale("log")
    for one_chart in charts:   
        plt.plot(*one_chart["matrix"], one_chart["character"])
        
generate_info_dict = lambda funcs_list, parameters: {func["func_name"]: func["func_cursor"](*parameters) for func in funcs_list}
```

# 1.1
Написать функцию, вычисляющую теоретические оценки значений характеристик графа в модели ER_n_p: ожидаемое (в среднем) количество рёбер, ожидаемая (в среднем) степень узла, оценка диаметра графа.
<https://arxiv.org/pdf/cond-mat/0407098.pdf>
<https://math.stackexchange.com/questions/501216/what-is-the-equation-for-the-average-path-length-in-a-random-graph>

```python
mean_count_edges_t = lambda n, p: p*n*(n-1)/2
mean_degrees_nodes_t = lambda n, p: p*n
diameter_graph_t = lambda n, p: (np.log(n))/(np.log(mean_degrees_nodes_t(n, p)))
coeff_claster_t = lambda n, p: mean_degrees_nodes_t(n, p)/n
mean_path_len_t = lambda n, p: (np.log(n) - γ)/np.log(mean_degrees_nodes_t(n, p)) + 0.5

funcs_list_t = [
    {"func_name": "mean_count_edges_t", "func_cursor": mean_count_edges_t},
    {"func_name": "mean_degrees_nodes_t", "func_cursor": mean_degrees_nodes_t},
    {"func_name": "diameter_graph_t", "func_cursor": diameter_graph_t},
    {"func_name": "coeff_claster_t", "func_cursor": coeff_claster_t},
    {"func_name": "mean_path_len_t", "func_cursor": mean_path_len_t}
]
```

```python
generate_info_dict(funcs_list_t, (1000, 0.5))
```

# 1.2. 
Написать функцию, вычисляющую характеристики реального графа (количество вершин, рёбер, максимальная степень вершины, средняя степень вершины, коэффициент кластеризации, диаметр, средняя длина пути).


```python
count_nodes_p = lambda G: len(G.nodes())
count_edges_p = lambda G: len(G.edges())
max_degree_nodes_p = lambda G: max(dict(G.degree()).values())
avg_degree_nodes_p = lambda G: sum(dict(G.degree()).values())/len(dict(G.degree()).values())
#coeff_claster_p = lambda G: avg_degree_nodes_p(G)/count_nodes_p(G)
coeff_claster_p = lambda G: nx.average_clustering(G)
diameter_graph_p = lambda G: nx.diameter(G)
avg_path_len_p = lambda G: nx.average_shortest_path_length(G)

funcs_list_p = [
    {"func_name": "count_nodes_p", "func_cursor": count_nodes_p},
    {"func_name": "count_edges_p", "func_cursor": count_edges_p},
    {"func_name": "max_degree_nodes_p", "func_cursor": max_degree_nodes_p},
    {"func_name": "avg_degree_nodes_p", "func_cursor": avg_degree_nodes_p},
    {"func_name": "coeff_claster_p", "func_cursor": coeff_claster_p},
    {"func_name": "diameter_graph_p", "func_cursor": diameter_graph_p},
    {"func_name": "avg_path_len_p", "func_cursor": avg_path_len_p}
]
```

```python
G = nx.gnp_random_graph(1000, 0.5)
generate_info_dict(funcs_list_p, (G,))
```

# 1.3. 
Написать функцию, расчитывающую отклонения расчётных значений средней длины пути и кластерного коэффициента от теоретических (слайд 24 лекции №2) в % от теоретических.

```python
def diffs_avgPathLen_coeffClaster(n, p):
    try:
        G = nx.gnp_random_graph(n, p)
        return {
            "diff_avgPathLen": abs(avg_path_len_p(G)-mean_path_len_t(n, p))/mean_path_len_t(n, p)*100,
            "diff_coeffClaster": abs(coeff_claster_p(G)-coeff_claster_t(n, p))/coeff_claster_t(n, p)*100,
        }
    except nx.NetworkXError: # happening, when the graph is not connected
        return {
            "diff_avgPathLen": np.nan,
            "diff_coeffClaster": abs(coeff_claster_p(G)-coeff_claster_t(n, p))/coeff_claster_t(n, p)*100,
        }
```

```python
chart_info_about_diff = [(diffs_avgPathLen_coeffClaster(1000, x), x,) for x in np.arange(0.0, 1.05, 0.05)]
chart_info_about_diff = list(filter(lambda x: not np.isnan(x[0]['diff_avgPathLen']+x[0]['diff_coeffClaster']), chart_info_about_diff))
```

```python
plot_chart("Diff avgPathLen", "p", "relative error (%)", False,
           {"matrix": [[x[1] for x in chart_info_about_diff], [x[0]['diff_avgPathLen'] for x in chart_info_about_diff]], "character": '.'})

plot_chart("Diff coeffClaster", "p", "relative error (%)", False,
           {"matrix": [[x[1] for x in chart_info_about_diff], [x[0]['diff_coeffClaster'] for x in chart_info_about_diff]], "character": '.'})
```

# 1.4. 
Написать функцию визуализации графа (вершины помечены). Подобрать раскладку графа, наиболее наглядно представляющую устройство графа (можно сделать раскладку аргументом функции).

```python
def plot_graph(G, draw_type, weight_name=None):
#     %matplotlib notebook
#     import matplotlib.pyplot as plt
    
    plt.figure()
    pos = nx.spring_layout(G)
    edges = G.edges()
    weights = None
    
    if weight_name:
        weights = [int(G[u][v][weight_name]) for u,v in edges]
        labels = nx.get_edge_attributes(G,weight_name)
        nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
        nx.draw_networkx(G, pos, width=weights);
    else:
        if draw_type == "circular":
            nx.draw_circular(G)
        elif draw_type == "planar":
            nx.draw_circular(G)
        elif draw_type == "random":
            nx.draw_circular(G)
        elif draw_type == "spectral":
            nx.draw_circular(G)
        elif draw_type == "spring":
            nx.draw_circular(G)
        else:
            nx.draw_networkx(G, pos)
```

# 1.5. 
Написать функцию, вычисляющую максимальный связный подграф данного графа (nx.strongly_connected_components). Если количество вершин связного подграфа не меньше 3/4 количества вершин данного графа, подграф выводится на экран и представляются его характеристики (функция 1.2).

```python
def max_connected_subgraph(G):
    max_connected_subgraph_G = G.subgraph(max(nx.connected_components(G), key=len))
    return {"check": True, "graph": max_connected_subgraph_G} if len(max_connected_subgraph_G.nodes()) >= 3/4*len(G.nodes()) else {"check": False, "graph": max_connected_subgraph_G}
```

```python
def chart_degrees_nodes(G, log=False):
    degrees = list(dict(G.degree()).values())
    frequency_data = sorted([(degree, frequency) for degree, frequency in dict(Counter(degrees)).items()], key=lambda x: x[0])
    plot_chart("Частотное распределение", "степень вершины", "количество вершин", log,
           {"matrix": [[x[0] for x in frequency_data], [x[1] for x in frequency_data]], "character": '.'})
```

# 1.7. 
Используя функцию nx.gnp_random_graph, сгенерировать случайный граф с параметрами
n = 100, p = 0,001. Применить к нему указанные выше функции. Увеличивая значение параметра p, найти его минимальное значение (с точностью до 1/(10n)), при котором проявляется эффект фазового перехода (связный подграф с 3/4 вершин). Распечатать этот подграф и его характеристики.

```python
sign = lambda x: copysign(1, x)

def binary_search_rec(count_nodes, left=0, right=1, counter=0):
    if abs((right+left)/2 - counter) < 0.1/count_nodes:
        return (right+left)/2
    counter = (right+left)/2
    G = nx.gnp_random_graph(count_nodes, counter)
    if max_connected_subgraph(G)['check']:
        G = None
        return binary_search_rec(count_nodes, left, counter, counter)
    else:
        G = None
        return binary_search_rec(count_nodes, counter, right, counter)
    
def binary_search_not_rec(count_nodes):
    left, right = 0, 1
    counter = 0
    while abs((right+left)/2 - counter) >= 0.1/count_nodes:
        counter = (right+left)/2
        G = nx.gnp_random_graph(count_nodes, counter)
        if max_connected_subgraph(G)['check']:
            right = counter
        else:
            left = counter
        del G
    return (right+left)/2

def bruteforce_search(count_nodes, start_value_p=0):
    step = 0.1
    prev_indicator = True
    prev_step = 0
    while True:
        indicator = max_connected_subgraph(nx.gnp_random_graph(count_nodes, start_value_p))['check']
        if indicator == prev_indicator:
            step = round(-0.1*step, ceil(np.log10(count_nodes))+1) if abs(0.1*step) > 0.1/count_nodes else round(-0.1*sign(step)/count_nodes, ceil(np.log10(count_nodes))+1)
            if prev_step == -step:
                break
            prev_step = step
            print(step)
            
        start_value_p += step
        prev_indicator = not indicator
    return start_value_p
    
def p_search(count_nodes, search_type="bi_s_nr", start_value_p=0, count_graph=1):
    if search_type == "bi_s_nr":
        p = np.mean(np.array([binary_search_not_rec(count_nodes) for _ in range(count_graph)]))
    elif search_type == "bi_s_r":
        p = np.mean(np.array([binary_search_rec(count_nodes) for _ in range(count_graph)]))
    elif search_type == "bf_s":
        p = np.mean(np.array([bruteforce_search(count_nodes, start_value_p) for _ in range(count_graph)]))
    return {'graph': max_connected_subgraph(nx.gnp_random_graph(count_nodes, p))['graph'], 'p': p}
```
