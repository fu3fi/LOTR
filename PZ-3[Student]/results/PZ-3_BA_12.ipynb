```python
import networkx as nx
from collections import Counter
from matplotlib import pylab as plt
import numpy as np
from scipy import stats
        
def plot_chart(title="", x="", y="", log=False, *charts):
    plt.figure(figsize=(8,6))
    plt.tight_layout()
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid()
    if log == True:
        plt.xscale("log")
        plt.yscale("log")
    for one_chart in charts:   
        plt.plot(*one_chart["matrix"], one_chart["character"])
        
generate_info_dict = lambda funcs_list, parameters: {func["func_name"]: func["func_cursor"](*parameters) for func in funcs_list}
```

```python
count_nodes_p = lambda G: len(G.nodes())
count_edges_p = lambda G: len(G.edges())
max_degree_nodes_p = lambda G: max(dict(G.degree()).values())
avg_degree_nodes_p = lambda G: sum(dict(G.degree()).values())/len(dict(G.degree()).values())
coeff_claster_p = lambda G: nx.average_clustering(G)
diameter_graph_p = lambda G: nx.diameter(G)

avg_path_len_p = lambda G: nx.average_shortest_path_length(G)

funcs_list_p = [
    {"func_name": "count_nodes_p", "func_cursor": count_nodes_p},
    {"func_name": "count_edges_p", "func_cursor": count_edges_p},
    {"func_name": "max_degree_nodes_p", "func_cursor": max_degree_nodes_p},
    {"func_name": "avg_degree_nodes_p", "func_cursor": avg_degree_nodes_p},
    {"func_name": "coeff_claster_p", "func_cursor": coeff_claster_p},
    {"func_name": "diameter_graph_p", "func_cursor": diameter_graph_p},
    {"func_name": "avg_path_len_p", "func_cursor": avg_path_len_p}
]

def get_coeff_cc_ba(m, work_type): # work up to 300, not inclusive
    if work_type == 'cc':
        chart_info_about_diff = [(np.log(x)**2/x, coeff_claster_p(nx.barabasi_albert_graph(x, m)),) for x in np.arange(300, 1050, 50)]
    elif work_type == 'apl':
        chart_info_about_diff = [(np.log(x)/np.log(np.log(x)), avg_path_len_p(nx.barabasi_albert_graph(x, m)),) for x in np.arange(300, 1050, 50)]
    else:
        class Get_Coeff(Exception):
            pass
        raise Get_Coeff('error of work type')

    slope, intercept, r_value, p_value, std_err = stats.linregress([x[0] for x in chart_info_about_diff], [x[1] for x in chart_info_about_diff])
    return {
        "slope": slope,
        "intercept": intercept,
    }

def diffs_avgPathLen_coeffClaster(n, m, coeff, G=None):
    if G == None:
        G = nx.barabasi_albert_graph(n, m)
    dapl = abs(avg_path_len_p(G)-(coeff['apl']['slope']*(np.log(n)/np.log(np.log(n)))+coeff['apl']['intercept']))/(coeff['apl']['slope']*(np.log(n)/np.log(np.log(n)))+coeff['apl']['intercept'])*100 \
        if abs(avg_path_len_p(G)-(coeff['apl']['slope']*(np.log(n)/np.log(np.log(n)))+coeff['apl']['intercept'])) != 0 else 0
    dcc = abs(coeff_claster_p(G)-(coeff['cc']['slope']*(np.log(n)**2)/n+coeff['cc']['intercept']))/(coeff['cc']['slope']*(np.log(n)**2)/n+coeff['cc']['intercept'])*100 \
        if abs(coeff_claster_p(G)-(coeff['cc']['slope']*(np.log(n)**2)/n+coeff['cc']['intercept'])) != 0 else 0
    return {
        "diff_avgPathLen": dapl,
        "diff_coeffClaster": dcc,
    }
```

```python
m = 10
coeff = dict()
coeff['cc'] = get_coeff_cc_ba(m, 'cc')
coeff['apl'] = get_coeff_cc_ba(m, 'apl')
chart_info_about_diff = [(diffs_avgPathLen_coeffClaster(x, m, coeff), x,) for x in np.arange(50, 1050, 50)]

plot_chart(f"Diff avgPathLen m={m}", "p", "relative error (%)", False,
           {"matrix": [[x[1] for x in chart_info_about_diff], [x[0]['diff_avgPathLen'] for x in chart_info_about_diff]], "character": '.'})

plot_chart(f"Diff coeffClaster m={m}", "p", "relative error (%)", False,
           {"matrix": [[x[1] for x in chart_info_about_diff], [x[0]['diff_coeffClaster'] for x in chart_info_about_diff]], "character": '.'})
```

# 3.1. 
Написать функцию, строящую график рспределения степеней графа в обычных и логарифмических координатах.

```python
def chart_degrees_nodes(G, log=False):
    degrees = list(dict(G.degree()).values())
    frequency_data = sorted([(degree, frequency) for degree, frequency in dict(Counter(degrees)).items()], key=lambda x: x[0])
    plot_chart("Частотное распределение", "степень вершины", "количество вершин", log,
           {"matrix": [[x[0] for x in frequency_data], [x[1] for x in frequency_data]], "character": '.'})
```

```python
chart_degrees_nodes(nx.barabasi_albert_graph(1000, 10), False)
chart_degrees_nodes(nx.barabasi_albert_graph(1000, 10), True)
```
